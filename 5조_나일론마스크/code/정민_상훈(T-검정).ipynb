```python
#to get consistent result
np.random.seed(1)
```

```python
#0
group1 = df[df['bankrupt'] == 0]
#1
group2 = df[df['bankrupt'] == 1]
variable_list = ["x1","x2","x3", "x4", "x5", "om", "ga", "gs", "ge", "croe", "cpb"]
for i in variable_list :

  # #perform t-test assuming equal variances 등분산성 t검정
  tTestResult= stats.ttest_ind(group1[f"{i}"], group2[f"{i}"])
  
  # #perform t-test NOT assuming equal variances 이분산성 t검정 (ga gs ge의 경우에만 활용)
  tTestResultDiffVar= stats.ttest_ind(group1[f"{i}"], group2[f"{i}"], equal_var=False)
  
  print(f"Variable {i} : The t-statistic and p-value assuming equal variances is %.3f and %.3f." % tTestResult)
  print(f"Variable {i} : The t-statistic and p-value not assuming equal variances is %.3f and %.3f" % tTestResultDiffVar)
#  x4, ga, gs, ge가 등분산성을 만족, 따라서 그 외의 것은 이분산성 t 검증 값을 봐야 한다
# 등분산성 검증에 따르면 모두 유의미하다
# x5, croe, cpb 의 경우 이분산성 t 검증 값에 유의미 하지 않다를 보여주고 있다. (나머지는 유의미)
# ga 의 경우 등분산성 t검증에 따르면 유의미하다고 나옴
# 결론 : x5, croe, cpb 의 경우 유의미 하지 않다
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import statsmodels.api as sm

for i in variable_list:
    
  model = ols(f'{i} ~ bankrupt', dat2).fit()
  print(f"{i}",anova_lm(model, typ=3))

# ga gs ge 관련 유의미성 부족
```

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

contingency = pd.crosstab(dat2.iloc[:,1],dat2.iloc[:,2])
contingency
dat2.iloc[:,1]
```

```python
chi, p, dof, expected = chi2_contingency(contingency)

print(f"chi 스퀘어 값: {chi}",
      f"p-value (0.05): {p}",
      f"자유도 수: {dof}",
      f"기대값: \n{pd.DataFrame(expected)}",
      f"측정값: \n{contingency}", sep = "\n" )
```
