```python
import pandas as pd
variable = pd.read_csv("/content/drive/MyDrive/부실기업예측 프로젝트/data/variable.csv")
print(variable.shape)
```

```python
# features = variable[['x1', 'x2', 'x3', 'x4', 'x5', 'om', 'ga', 'gs', 'ge', 'croe', 'cpb']]
features = variable[['x1', 'x2', 'x3', 'x4',  'om', 'ga', 'gs', 'ge']]
bankrupt = variable['bankrupt']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(features, bankrupt,stratify=bankrupt,random_state=80,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, confusion_matrix
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
print(model.score(X_train, y_train))
```

```python
print(model.score(X_test, y_test))
```

```python
test_predict = model.predict(X_test)
roc_auc_score(y_test, test_predict)
```

```python
# 재현율
recall_score(y_test, test_predict)
```

```python
confusion_matrix(y_test, test_predict)
```

```python
print(model.coef_)
```

# 오버샘플링


```python
from sklearn.datasets import make_classification
from sklearn.decomposition import PCA
from imblearn.over_sampling import SMOTE
from collections import Counter
from matplotlib import pyplot


```

```python
X_resampled, y_resampled = SMOTE(random_state=0).fit_resample(X_train, y_train)
```

```python

print('After OverSampling, the shape of train_X: {}'.format(X_resampled.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_resampled.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_resampled==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_resampled==0)))
```

```python
model_2 = LogisticRegression()
model_2.fit(X_resampled, y_resampled)
```

```python
model_2.score(X_resampled, y_resampled)
```

```python
test_predict_resampled = model_2.predict(X_test)
roc_auc_score(y_test, test_predict_resampled)
```

```python
# 정확도
accuracy_score(y_test, test_predict_resampled)
```

```python
# 재현율
recall_score(y_test, test_predict_resampled)
```

```python
confusion_matrix(y_test, test_predict_resampled)
```
