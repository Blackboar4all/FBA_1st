```python
#데이터불러오기
df = pd.read_csv("/content/drive/My Drive/부실기업예측 프로젝트/data/variable.csv", encoding="utf-8")
```

```python
df.head()
```

```python
#변수선정
X = df[['x1','x2','x3','x4','x5','om','ga','gs','ge','croe','cpb']]
y = df['bankrupt']
print('Data View')
display(pd.concat([pd.DataFrame(y, columns=["Bankrupt"]), pd.DataFrame(X, columns=X.columns)], axis=1).head())
```

```python
alpha_weight = 0.5
fit = Lasso(alpha=alpha_weight, fit_intercept=True, normalize=True, random_state=123).fit(X, y)
pd.DataFrame(np.hstack([fit.intercept_, fit.coef_]), columns=['alpha = {}'.format(alpha_weight)])
```

```python
#Ridge
alpha_weight = 0.5
fit = Ridge(alpha=alpha_weight, fit_intercept=True, normalize=True, random_state=123).fit(X, y)
pd.DataFrame(np.hstack([fit.intercept_, fit.coef_]), columns=['alpha = {}'.format(alpha_weight)])
```

```python
#Lasso
alpha_weight = 0.01
fit = Lasso(alpha=alpha_weight, fit_intercept=True, normalize=True, random_state=123).fit(X, y)
pd.DataFrame(np.hstack([fit.intercept_, fit.coef_]), columns=['alpha = {}'.format(alpha_weight)])
```

```python
# Elastic
fit = Ridge(alpha=0.5, fit_intercept=True, normalize=True, random_state=123).fit(X, y)
pred_tr = fit.predict(X)
pred_te = fit.predict(X)
```
