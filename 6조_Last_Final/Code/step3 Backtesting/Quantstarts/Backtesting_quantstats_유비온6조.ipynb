```python
for stock in stocks:
  for k in stock.keys():
    stock = qs.utils.download_returns(stock[k])
    metrics_temp = qs.reports.metrics(stock, mode='basic', display=False) #full
    metrics = metrics_temp.T
    metrics['name'] = k
    df = df.append(metrics[myColumns], ignore_index=True)
```
